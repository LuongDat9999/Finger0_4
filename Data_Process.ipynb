```python
%pip install opencv-python mediapipe Pillow

import os
import cv2
import mediapipe as mp
from PIL import Image
import numpy as np

# Khi to MediaPipe Hands
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=True, max_num_hands=1, min_detection_confidence=0.5)

# Th mc gc cha cc th mc con
base_dir = 'RawData'
output_dir = 'Clean_data'

# To th mc u ra nu n cha tn ti
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Lp qua tng th mc con
for x in range(5):
    input_dir = os.path.join(base_dir, str(x))
    output_subdir = os.path.join(output_dir, str(x))
    
    # To th mc con u ra nu n cha tn ti
    if not os.path.exists(output_subdir):
        os.makedirs(output_subdir)
    
    # Lit k tt c cc tp trong th mc u vo
    files = os.listdir(input_dir)
    
    print(f"ang x l th mc: {input_dir} vi {len(files)} tp")
    
    # X l tng tp trong th mc con
    for y, file_name in enumerate(files):
        # To ng dn tp y 
        file_path = os.path.normpath(os.path.join(input_dir, file_name))
        
        try:
            # c hnh nh bng Pillow
            pil_image = Image.open(file_path).convert('RGB')
            image = np.array(pil_image)
            
            # Chuyn i hnh nh RGB sang BGR (nh dng OpenCV)
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
            
            # Chuyn i hnh nh BGR sang RGB
            image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            
            # X l hnh nh v tm tay
            results = hands.process(image_rgb)
            
            
            if results.multi_hand_landmarks:
                for hand_landmarks in results.multi_hand_landmarks:
                    # Ly hp gii hn ca tay
                    x_min = min([landmark.x for landmark in hand_landmarks.landmark])
                    y_min = min([landmark.y for landmark in hand_landmarks.landmark])
                    x_max = max([landmark.x for landmark in hand_landmarks.landmark])
                    y_max = max([landmark.y for landmark in hand_landmarks.landmark])
                    
                    # Chuyn i ta  chun ha sang ta  pixel
                    h, w, _ = image.shape
                    x_min = int(x_min * w)
                    y_min = int(y_min * h)
                    x_max = int(x_max * w)
                    y_max = int(y_max * h)
                    
                    # Tng kch thc hp gii hn ln 10%
                    x_margin = int((x_max - x_min) * 0.1)
                    y_margin = int((y_max - y_min) * 0.1)
                    x_min = max(0, x_min - x_margin)
                    y_min = max(0, y_min - y_margin)
                    x_max = min(w, x_max + x_margin)
                    y_max = min(h, y_max + y_margin)
                    
                    #
```
